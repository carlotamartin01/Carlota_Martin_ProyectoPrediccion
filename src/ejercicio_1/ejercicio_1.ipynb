```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python

########################
# IMPORTAMOS LOS DATOS #
########################

data = pd.read_csv("../../data/data_cleaned.csv")
data.head()
```

```python

############################
# CREAMOS COLUMNA DE GASTO #
############################


data["Gasto_total"] = (data["MntWines"] + 
                          data["MntFruits"] + 
                          data["MntMeatProducts"] + 
                          data["MntFishProducts"] + 
                          data["MntSweetProducts"] + 
                          data["MntGoldProds"])

data[["Income", "Gasto_total"]].head()
```

```python
########################
#  INGRESOS VS GASTOS  #
########################
sns.scatterplot(x=data["Income"], y=data["Gasto_total"])
plt.title("RelaciÃ³n entre Ingresos y Gastos")
plt.xlabel("Ingresos")
plt.ylabel("Gastos")
plt.show()
```

```python
#############################
#  CORR INGRESOS VS GASTOS  #
#############################

correlacion = data["Income"].corr(data["Gasto_total"])
print(f"La correlaciÃ³n entre ingresos y gastos es de:{correlacion: }")
```
